```python
class RecordComparator():
    def __init__(self, key, typ):
        self.key = key
        self.typ = typ
        
    def compare(self, rec1, rec2):
        typOrder = {'asc':1, 'desc':-1}
        if self.key not in rec1:
            val1 = 0
        else:
            val1 = rec1[self.key]
        if self.key not in rec2:
            val2 = 0
        else:
            val2 = rec2[self.key]
        if val1 < val2:
            return typOrder[self.typ]*-1
        elif val1 > val2:
            return typOrder[self.typ]*1
        else: 
            return 0
```
